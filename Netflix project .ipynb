```python
# CODE
df.loc[df['show_id'] == 's5542', 'rating'] = 'TV-MA'
df.loc[df['show_id'] == 's5795', 'rating'] = 'TV-MA'
df.loc[df['show_id'] == 's5814', 'rating'] = 'TV-MA'
```

```python
# CODE
df.loc[df['show_id'] == 's5542', 'duration'] = '74 min'
df.loc[df['show_id'] == 's5795', 'duration'] = '84 min'
df.loc[df['show_id'] == 's5814', 'duration'] = '66 min'
```

### Distribution of TV shows and Movies

```python
directors= df['director'].str.split(',' , expand=True).stack()
```

```python
directors  #We have a list of directors 
```

```python
directors=directors.to_frame()
directors
```

```python
directors.columns=['director']
directors
```

```python
temp_D=directors.groupby(['director']).size().reset_index(name='count')
```

```python
temp_D
```

```python
temp_D=temp_D[temp_D.director != 'Not Specified']
```

```python
temp_D=temp_D.sort_values(by='count' , ascending =False)
temp_D= temp_D.head(10)
temp_D
```

######  we have extracted the names of directors from the data and their total count 

```python
sns.barplot(temp_D , x='count', y='director' )
plt.show()
```

- The above barplot made with the help of seaborn shows the top 10 directors on netflix.

## Which country has most no. of content

```python
country=df['country'].str.split(',' , expand=True).stack()
country=country.to_frame()
```

```python
country.columns=['Country']
country
```

```python
temp_P=country.groupby(['Country']).size().reset_index(name='count')
temp_P=temp_P.sort_values(by='count' , ascending=False)
temp_P=temp_P[temp_P.Country != 'Unknown']
temp_P=temp_P.head(10)
temp_P
```

## Popular Genra Analysis

```python
sns.countplot(data=df , y='listed_in' , order=df['listed_in'].value_counts().index[:20])
```

- Here the show/movie has more than one genre so now we will find the separate genre analysis to get to know more about types of genre which is popular on netflix

```python
genre=df['listed_in'].str.split(',' , expand=True).stack()
genre=genre.to_frame()
genre.columns=['Genre']
genre
```

```python
df3=genre.groupby(['Genre']).size().reset_index(name='count')
df3=df3.sort_values(by='count' , ascending=False)
df3=df3.head(20)
df3
```

```python
sns.barplot(df3 , x='count' , y='Genre')
plt.show()
```
